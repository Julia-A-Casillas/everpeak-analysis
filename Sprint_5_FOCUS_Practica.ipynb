JOINS

```python
clientes = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "nombre": ["Ana", "Luis", "María", "Carlos", "Sofía"]
})
clientes
```

```python
compras = pd.DataFrame({
    "id_cliente": [1, 2, 2, 4, 6],
    "producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Tablet"]
})
compras
```

```python
clientes.merge(compras,how='inner', on='id_cliente')
```

```python
clientes.merge(compras,how='inner', left_on='id', right_on='id_cliente')
```

```python
clientes.merge(compras,how='left', left_on='id', right_on='id_cliente')
```

```python
clientes.merge(compras,how='right', left_on='id', right_on='id_cliente')
```

```python
clientes.merge(compras,how='outer', left_on='id', right_on='id_cliente')
```
